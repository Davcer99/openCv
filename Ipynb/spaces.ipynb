```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Images/cat.jpg')# default is BGR
cv.imshow('Cat', img)
```

GR to gray

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
""" cv.imshow('Gray', gray)
 """
#BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
""" cv.imshow('HSV', hsv)
 """
#BGR to LAB
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
""" cv.imshow('LAB', lab) """
```

GR to RGB

```python
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
"""cv.imshow('RGB', rgb)"""
```

```python
plt.imshow(rgb)
plt.show()
```
